```
from youtube_dataset_creator import youtube_search
from youtube_dataset_creator import youtube_database
from youtube_dataset_creator import youtube_download
```

```
yts = youtube_search.YouTubeSearch(project_dir)
```

```
celebrity_name = 'michael stuhlbarg'
search_term = 'interview'
filename = celebrity_name.replace(' ','_') + '_' + search_term
```

```
results = yts.search(celebrity_name, search_term)
```
